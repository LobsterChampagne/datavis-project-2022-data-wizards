```python
import pandas as pd
```

```python
df = pd.read_csv('Data/all_streams.csv', index_col=0)
```

```python
df.head()
```

```python
a = pd.DataFrame(df.groupby('release_year')['release_year'].agg('count'))
b = a.index.values
c = pd.DataFrame((b,a['release_year'])).T
c.columns = ['release_year', 'count']
c
```

```python
c.to_csv('Data/release_year.csv', index=False)
```
